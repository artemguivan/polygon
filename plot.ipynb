```python
import pathlib
from typing import List
import seaborn as sns
import matplotlib.pyplot as plt

def get_test_and_time(filename: pathlib):
    data = []
    with open(filename) as f:
        data.append(f.read())

    data = "".join(data).split("\n")[:-1]
    return data

sep_data = get_test_and_time("hello.txt")

def get_data_for_graph(data: List[str]):
    count_set_time_dict = {}
    for line in data:
        count_set, time = line.split()
        count_set_time_dict[int(count_set)] = int(time)
    return count_set_time_dict
data_for_graph = get_data_for_graph(sep_data)

sns.lineplot(data=data_for_graph, y=data_for_graph.values(), x=data_for_graph.keys());

plt.xlabel("count of sets")   
plt.ylabel("microseconds")   
plt.title("Plot (Universe - 5)")
```
